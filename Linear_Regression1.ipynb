```python
X= [-2,-1,1,3]
Y= [-2.112,-0.945,3.102,6.089]
n= len(X)
m= len(Y)
```

```python
mean_x = sum(X)/n
mean_y = sum(Y)/m
print(mean_x)
print(mean_y)
```

```python
d1 = mean_x
d2 = mean_y
```

```python
dx = (X - d1 for x in X)
dy = (Y - d2 for y in Y)
```

```python
dx_values = [x - d1 for x in X]
dy_values = [y - d2 for y in Y]

numerator = sum([dx * dy for dx, dy in zip(dx_values, dy_values)])
denominator = sum([dx_val**2 for dx_val in dx_values])
m = numerator / denominator
print(m)
```

```python
b= mean_y - m*mean_x
print(b)
```
